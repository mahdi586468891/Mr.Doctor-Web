```python
import os,shutil,random
from os import path
```

```python
cancer_list = os.listdir(r"D:\Parsia Works\python\Project\SMAmodel\datasets")
cancer_list
```

```python
os.mkdir("new_datasets")

test_dir = r"new_datasets\\test"
train_dir = r"new_datasets\\train"

os.mkdir(test_dir)
os.mkdir(train_dir)

for _cancer in cancer_list :
    os.mkdir(path.join(train_dir,_cancer))
    os.mkdir(path.join(test_dir,_cancer))

```

```python
for _cancer in cancer_list :
    print(f"******* {_cancer} *******")

    _all_class = os.listdir(path.join("datasets",_cancer,"test"))

    for _type in ["test","train"]:
        img_paths = []
        print(f"    ******* {_cancer}.{_type} *******")


        for _class in _all_class:
            print(f"        ******* {_cancer}.{_type}.{_class} *******")


            l = os.listdir(path.join("datasets",_cancer,_type,_class))
            for _l in l :
                img_paths.append(f"{_class}#{_l}")

        random.shuffle(img_paths)

        for item in img_paths :
            print(item)
            _class_,name = item.split("#")
            shutil.move(path.join("datasets",_cancer,_type,_class_,name)
                        ,path.join("new_datasets",_type,_cancer,name)
            )
        
```
