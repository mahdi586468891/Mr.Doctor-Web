```python
import os
import shutil
import numpy as np
```

```python

for path in os.listdir("Training"):
    os.mkdir(f"Testing\\{path}")
    print(path)
```

```python
TEST_SIZE = 0.2
```

```python
for _classes in os.listdir("Training"):
    print(_classes)
    images_path = os.listdir(os.path.join("Training",_classes))
    np.random.shuffle(images_path)

    test_first_index = int(TEST_SIZE*len(images_path))

    most_to_go_to_test = images_path[:test_first_index]
    l = len(images_path)
    i = 1
    for item in most_to_go_to_test :
        os.system("cls")
        print(i/l)
        _tr = os.path.join("Training",_classes,item)
        _te = os.path.join("Testing",_classes,item)

        shutil.move(_tr,_te)
        i += 1

    
```
